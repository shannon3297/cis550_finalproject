```python
# imports
from google.colab import drive
import pandas as pd

# mount to drive
drive.flush_and_unmount()
drive.mount('/content/drive', force_remount=True)
```

```python
path = '/content/drive/MyDrive/F21/CIS550/CIS 550 Final Project (SHARED)/WSJ Articles/'
df = pd.read_csv(path + 'russell1000stocks_final.csv')
company_names = df['company_name']
```

```python
import re
stopwords = [' Technologies ', ' panies ', ' International ', ' Plc ', ' Ltd ', ' pany ', ' Group ', ' Airlines ',' oration ', ' plc ', ' PLC ', ' Limited ', ' Therapeutics ', ' orporated ', ' Holdings ']  # trim and get rid of - if it's the last thing
cleaned_company_names = []
for company in company_names:
  curr_company = ' '.join([word for word in company.split() if word not in stopwords])
  if re.match(r'(.*?)- *$', curr_company):
    curr_company = re.sub('-', '', curr_company)
  curr_company = curr_company.strip()
  cleaned_company_names.append(curr_company)
cleaned_company_names
```

```python
df['company_name'] = cleaned_company_names
path = '/content/drive/MyDrive/F21/CIS550/CIS 550 Final Project (SHARED)/WSJ Articles/'
df.to_csv(path + 'russell1000stocks_finalCleaned.csv', index=False)
```

```python
og_caps = df['market_cap']
cleaned_caps = []
for cap in og_caps:
  cap = cap.strip('$').strip()
  cap = re.sub(",", "", cap)
  cap = float(cap)
  cleaned_caps.append(cap)
```

```python
df['market_cap'] = cleaned_caps
df.to_csv(path + 'russell1000stocks_finalCleaned.csv', index=False)
```
